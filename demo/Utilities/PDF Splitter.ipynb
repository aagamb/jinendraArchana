```python
import os
import PyPDF2

# Function to split and save PDFs
def split_pdf(input_pdf_path, output_dir, books):
    # Create output directory if not exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir, exist_ok=True)

    # Open the PDF
    with open(input_pdf_path, "rb") as pdf_file:
        reader = PyPDF2.PdfReader(pdf_file)
        total_pages = len(reader.pages)

        # Iterate through books and extract pages
        for i in range(len(books)):
            book = books[i]
            # book_name = book["name"].replace(" ", "_")  # Convert spaces to underscores
            book_name = book["name"]
            start_page = book["pgNum"] - 1  # Convert to zero-based index
            end_page = books[i + 1]["pgNum"] - 2 if i + 1 < len(books) else total_pages - 1  # Last page for book
            
            # Create a new PDF writer
            writer = PyPDF2.PdfWriter()
            for page_num in range(start_page, end_page + 1):
                writer.add_page(reader.pages[page_num])

            # Save the split PDF
            output_pdf_path = os.path.join(output_dir, f"{book_name}.pdf")
            with open(output_pdf_path, "wb") as output_pdf:
                writer.write(output_pdf)
            
            print(f"Created: {output_pdf_path}")

# Example usage
input_pdf_path = "Jinendra Archana.pdf"  # Change to your actual PDF file
output_dir = "temp"
split_pdf(input_pdf_path, output_dir, books)
```
